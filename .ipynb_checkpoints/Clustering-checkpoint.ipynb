## Data preparation

### Subtask:
Prepare the data for clustering by encoding categorical features and scaling numerical features.


**Reasoning**:
Separate categorical and numerical features, then one-hot encode the categorical features and standardize the numerical features.



```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
import pandas as pd
import numpy as np

# Identify categorical and numerical features
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
numerical_cols = df.select_dtypes(include=['number']).columns.tolist()

# Create transformers for categorical and numerical features
numerical_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Combine transformers using ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])


# Fit and transform the data
df_processed = preprocessor.fit_transform(df)

# Get feature names after one-hot encoding
feature_names = (
    numerical_cols + list(preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_cols))
)

# Convert the processed data back to a DataFrame
df_scaled = pd.DataFrame(df_processed, columns=feature_names)

# Display first few rows and info
display(df_scaled.head())
print(df_scaled.info())
```

## Data clustering

### Subtask:
Perform K-means clustering on the preprocessed data.


**Reasoning**:
Perform K-means clustering on the preprocessed data, evaluate the clustering quality using the Silhouette score, iterate on the n_clusters parameter to find the optimal number of clusters, and plot the silhouette scores.



```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Experiment with different values for n_clusters
silhouette_scores = []
for n_clusters in range(2, 20):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(df_scaled)
    silhouette_avg = silhouette_score(df_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    print(f"For n_clusters = {n_clusters}, the average silhouette_score is : {silhouette_avg}")

# Plot the silhouette scores
plt.figure(figsize=(8, 6))
plt.plot(range(2, 20), silhouette_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. Number of Clusters')
plt.show()

# Find the optimal number of clusters based on the highest Silhouette score
optimal_n_clusters = range(3, 11)[silhouette_scores.index(max(silhouette_scores))]
print(f"Optimal number of clusters: {optimal_n_clusters}")

# Fit the KMeans model with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_n_clusters, random_state=42, n_init=10)
df_scaled['cluster_label'] = kmeans.fit_predict(df_scaled)

display(df_scaled.head())
```

## Data visualization

### Subtask:
Visualize the clusters to understand the separation between them.


**Reasoning**:
Visualize the clusters using scatter plots of the first two principal components and histograms/box plots of relevant features.



```python
import matplotlib.pyplot as plt

# Scatter plot of the first two principal components (or other relevant features)
plt.figure(figsize=(8, 6))
plt.scatter(df_scaled['Total day minutes'], df_scaled['Total eve minutes'], c=df_scaled['cluster_label'], cmap='viridis')
plt.xlabel('Total Day Minutes')
plt.ylabel('Total Eve Minutes')
plt.title('Clusters based on Total Day Minutes and Total Eve Minutes')
plt.colorbar(label='Cluster Label')
plt.show()

# Histograms/box plots for relevant features within each cluster
features_to_plot = ['Total day minutes', 'Total eve minutes', 'Total night minutes', 'Total intl minutes', 'Customer service calls']

plt.figure(figsize=(15, 10))
for i, feature in enumerate(features_to_plot):
  plt.subplot(2, 3, i + 1)
  for cluster in df_scaled['cluster_label'].unique():
    plt.hist(df_scaled[df_scaled['cluster_label'] == cluster][feature], bins=20, alpha=0.6, label=f'Cluster {cluster}')
  plt.xlabel(feature)
  plt.ylabel('Frequency')
  plt.title(f'Distribution of {feature} across Clusters')
  plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 10))
for i, feature in enumerate(features_to_plot):
  plt.subplot(2, 3, i + 1)
  for cluster in df_scaled['cluster_label'].unique():
    plt.boxplot(df_scaled[df_scaled['cluster_label'] == cluster][feature], positions=[cluster], widths=0.6, showfliers=False, patch_artist=True)
  plt.xlabel('Cluster')
  plt.ylabel(feature)
  plt.title(f'Boxplot of {feature} across Clusters')
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# Calculate the mean of the relevant features for each cluster
cluster_means = df_scaled.groupby('cluster_label')[features_to_plot].mean().reset_index()

# Melt the DataFrame to a long format for Plotly
cluster_means_melted = cluster_means.melt(
    id_vars='cluster_label',
    var_name='Feature',
    value_name='Mean Value'
)

# Create the radar plot
fig = px.line_polar(
    cluster_means_melted,
    r='Mean Value',
    theta='Feature',
    color='cluster_label',
    line_close=True,
    title='Radar Plot of Cluster Characteristics'
)
fig.show()
```
